```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df
```

```python
# histogram of age distribution

fig = sns.histplot(df, x='age', color='red', edgecolor='black', kde=True, bins=70)
fig.set_title("age distribution of titanic passengers")
fig.set_xlabel('age', color='green')
fig.set_ylabel('Frequency', color='green')
```

```python
# count of survival based on gender

fig = sns.countplot(df, x='survived', hue='sex')
fig.set_title("Count of titanic passengers by survival status")
fig.set_xlabel('Survival status', color='green')
fig.set_ylabel('Count', color='green')
```

```python
fig = sns.barplot(df, x='class', y='fare', hue='sex')
fig.set_title("Average fare by class and gender")
fig.set_xlabel('class')
fig.set_ylabel('average fare')
```

```python
fig = sns.countplot(df, x='class')
fig.set_title("Count of passenger by class")
fig.set_xlabel('class')
fig.set_ylabel('Count')
```

```python
fig = sns.pairplot(df[['age', 'survived', 'pclass', 'fare']], hue='survived')
fig
```
